```python
def cleaning(state, state_name):
    '''Performing data cleaning for each state'''
    important_attributes = list(state.columns)
    cleaned_state = {}
    for i in important_attributes:
        cleaned_state[i] = state[i].dropna()

    return {state_name: cleaned_state}

cleaned_states = {}


# Loop through the list of states
for state, state_name in zip(states, state_names):
    cleaned_states.update(cleaning(state, state_name))






```
